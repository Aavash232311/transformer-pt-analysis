```python
pred = [
    [5, 7, 0, 7],
    [3, 0, 7, 7],
    [1, 3, 4, 7],
] 

y = [
    [7, 7, 0, 7], 
    [3, 0, 7, 7],  
    [1, 3, 4, 7],  
]
correct = 0
total = 0
for pred_row, y_row in zip(pred, y):
    pred_skipped = pred_row[1:]  
    y_skipped    = y_row[1:]     

    for p, t in zip(pred_skipped, y_skipped):
        if p == t:
            correct += 1
        total += 1

print(f"Accuracy: {correct / total:.2%}") 
```

In the above code we are skipping the first column cause we can't predict what added to get the first element.

## Let's explore the masking here.

We have a restricted masking here so that the model cannot memorize the long sequence, and patterns.

```python
import torch

T = 4
mask = torch.full((T, T), float('-inf'))

for i in range(T):
    for j in range(max(0, i-2), i+1):
        mask[i, j] = 0.0

print(mask)
```
