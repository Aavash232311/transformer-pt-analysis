## Understanding of PCA analysis

```python
import torch
import matplotlib.pyplot as plt    

torch.manual_seed(11)
x = torch.randint(0, 4, (4,)).tolist()
y = torch.randint(0, 4, (4,)).tolist()

plt.plot(x, y, 'o')
plt.title("Plot of random data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Average Point

```python
x_a = sum(x) / len(x)
y_a = sum(y) / len(y)

plt.scatter(x, y, color='blue', label='Data')

plt.scatter(x_a, y_a, color='red', marker='.', s=200, label="Average")
plt.xlabel('X')
plt.ylabel('Y')

space = 8

plt.xlim(x_a - space, y_a  + space)
plt.ylim(x_a - space, y_a + space)
plt.title('Appending average of data')
plt.legend() # for showing labels
plt.grid(True)
plt.show()
```

### Trying to adjust a line to it


We can draw a random like that goes through the origin. 


Then we rotate the line till it fits the data as well as it can.


How can we conclude if a fit is good or not?


PCA projects the data onto it and it can either calculate the distance from the data to the line and try to find the line and try to find the line that minimizes distance. 


Or it can try to find the line that maximizes the distance from the projected points to the origin.



<b> Okay, so we need to project the data onto the line and measure the distance from the origin. <b/>


Now we rotate the line, project the data onto the line and sum the square distances we do until we come up with the line with the largest sum of squared distances between the projected points and the origin. 

Which is called `PC1`

`PC2` is the line that is perpendicular to the line `PC1

<hr />`



Let's draw a line, passing from origin to this average.


y2 - y1 = m (x2 - x1)


y2 - 1.25 = 1 (x2 - 1.75)


y2 = x2 - 1.75 + 1.25


y2 = x2 - 0.5 (Assuming this is PC1)

<hr />

<b> PC2 </b>

m1 = 1


m1.m2 = -1


m2 = -1/1 = -1


P (1.75, 1.25) passing point for line 2


y2 - y1 = m (x2 - x1)


y2 - 1.25 = -1 (x2 - 1.75)


y2 = -1x2 + 1.75 + 1.25


y2 = -x2 + 3

```python
import numpy as np
x_a = sum(x) / len(x)
y_a = sum(y) / len(y)
print(f"Avg (x {x_a}: , y: {y_a})")

plt.scatter(x, y, color='blue', label='Data')

plt.scatter(x_a, y_a, color='red', marker='.', s=200, label="Average")
plt.xlabel('X')
plt.ylabel('Y')

m = 1
c = -0.5
x_line = np.linspace(min(x)-1, max(y)+1, 150)
y_line = m * x_line + c
plt.plot(x_line, y_line, color='purple', label='PC1', linestyle="--") 


m2 = -1
c2 = 3
x2_line = np.linspace((min(x) - 1), max(y) + 1, 150)
y2_line = m2 * x2_line + c2
plt.plot(x2_line, y2_line, color='orange', label='PC2', linestyle="--") 


space = 8

plt.xlim(x_a - space, y_a  + space)
plt.ylim(x_a - space, y_a + space)
plt.title('Appending average of data')
plt.legend() 
plt.grid(True)
plt.show()
```
